```python
sz = 224
arch=resnet50
```

Because I am using a pretrained model from ImageNet, I will be using the same size they used to train their model (224). Altering it will force my model to re-learn everything - which I don't want. I will also use ResNet-50 as an architecture. 

ResNet-50 is a convolutional neural network that is trained on more than a million images from the ImageNet database. The network is 50 layers deep and can classify images into 1000 object categories.

```python
data = ImageClassifierData.from_paths(PATH, tfms = tfms_from_model(arch, sz))
learn = ConvLearner.pretrained(arch, data)
learn.fit(0.01, 3)
```

One thing I like about Fastai is its simplicity. Indeed, these three lines of code were able to build a model which has a prediction accuracy of 78%:

1. I created a data object, using ImageClassifierData, which encapsulates DataLoaders and Datasets for training, validation and test sets. It has many methods but the suitable one in this case is '.from_paths' as I am using folders to store the images. Training, validation and test set may come in different aspects such as 'csv' files. In that case, this method will not work
2. I created a learn object, using ConvLearner, which trains a chosen supported convnet model. Because I am using a pretrained model I passed to the 'pretrained' method two arguments, the architecture (ResNet-50) and the data object
3. I passed two arguments to the 'fit' method, the learning rate and the number of epochs

I chose 0.01 as a first learning rate because I have noticed that it was widely used. It's a good start to see how the model performs

The model got 78& accuracy as a start, using only three lines of code. Time to see if we can do better.

The results actually got worse. Next, I will use data augmentation. The main idea behind data augmentation is to take each image and modify it a little bit to "lure" the model and makes it think it's a different image. For example, it can take an image and flip it a little bit, rotate it, alter its luminosity...etc, to make it seem as a different image. It, generally speaking, creates a new data from an existing data. This article explains it better:
https://medium.com/nanonets/how-to-use-deep-learning-when-you-have-limited-data-part-2-data-augmentation-c26971dc8ced

```python
tfms = tfms_from_model(resnet34, sz, aug_tfms=transforms_side_on, max_zoom=1.1)
data = ImageClassifierData.from_paths(PATH, bs = 2, tfms = tfms)
learn = ConvLearner.pretrained(arch, data)
```

New argument: 'aug_tfms' - which standes for augmented transformation = transforms_side_on. Transforms Side On is specific to ImageNet-like images.

```python
def get_augs():
    data = ImageClassifierData.from_paths(PATH, bs=2, tfms=tfms, num_workers=1)
    x,_ = next(iter(data.aug_dl))
    return data.trn_ds.denorm(x)[1]
```

```python
ims = np.stack([get_augs() for i in range(9)])
```

```python
learn.fit(0.006, 3, cycle_len=1)
```

Cycle_len = 1 

This enables stochastic gradient descent with restarts (SGDR). The idea is as we get closer to the spot with the minimal loss we will start by gradually decreasing the learning rate. This technique is called learning rate annealing.

However, sometimes, the model will find itself in a part of the weight space that is a bit volatile (small changes to the weights can, for example, result in big changes to the loss). The idea is to allow it to find parts of the weight space that are accurate and more stable by increasing the learning rate and forcing the model to 'jump' to a different part of the wieght space (these are the 'restarts' in SGDR)

The number of epochs between resetting the learning rate is set by cycle_len, and the number of times this happens is referred to as the number of cycles.

```python
learn.fit(0.0035, 3, cycle_len=1, cycle_mult=2)
```

Cycle_multi = 2

This multiplies the length of the cycle after each cycle, as illustrated below. This allows the model to spend a bit more time on the spot looking for the optimal learning rate. Here, after the .fit method, the number of epochs is 7, because it was doubled every time we ran a cycle (1+2+4 = 7).

```python
learn.sched.plot_lr()
```

```python
log_preds,y = learn.TTA()
probs = np.mean(np.exp(log_preds),0)
```

The model now has 87% accuracy, which is not bad at all. But there's another thing that I can do to improve it a bit, and it is called TTA (Test Time Augmentation). It basically takes four random augmanted images + the original one, makes a prediction on each one of them and averages the results. Important to understand that this step is done after the model is trained, therefore, on the validation/test set. Predictions are generally expressed in log, therefore, exp will cancel it, hence np.exp.

The variable y represents the number (0 or 1) assigned to each meteorite class. Here, meteorite images that have a prediction score less than 0.5 are classified as iron meteorites, and those who have a prediction score between 0.5 and 1 are classified as stone meteorites. Therefore, if a meteorite gets a score of 0 it will indicate that the model correctly classified an irone meteorite perfectly. If a meteorite gets a score close to 0.5, it will indicate that the model had difficulties classifying it.
